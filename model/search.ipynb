```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Email Text'], df['Email Type'], test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# Define a list of pipelines for different models
pipelines = [
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', LogisticRegression(max_iter=1000))
    ]),
    # Pipeline([
    #     ('tfidf', TfidfVectorizer(stop_words='english')),
    #     ('classifier', SVC(random_state=42))
    # ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', RandomForestClassifier(random_state=42))
    ]),
    # Pipeline([
    #     ('tfidf', TfidfVectorizer(stop_words='english')),
    #     ('classifier', MLPClassifier(hidden_layer_sizes=(100,), random_state=42))
    # ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', RidgeClassifier())
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', SGDClassifier(max_iter=1000, tol=1e-3))
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', LinearSVC())
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', DecisionTreeClassifier())
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', ExtraTreesClassifier(n_estimators=100))
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', GradientBoostingClassifier())
    ]),
    Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('classifier', MultinomialNB())
    ]),
    # Pipeline([
    #     ('tfidf', TfidfVectorizer(stop_words='english')),
    #     ('classifier', KNeighborsClassifier(n_neighbors=3))
    # ])
]
```

```python
# Iterate over each pipeline, train, and evaluate
for pipeline in pipelines:
    print(f"trainning: {pipeline['classifier'].__class__.__name__}")
    # Fit the pipeline on the training data
    start_time = time.time()
    pipeline.fit(X_train, y_train)
    fit_time = time.time() - start_time
    print(f"Fit time: {fit_time:.4f} seconds")
    print(f"predicting: {pipeline['classifier'].__class__.__name__}")
    # Predict on the test set
    start_time = time.time()
    y_pred = pipeline.predict(X_test)
    predict_time = time.time() - start_time
    print(f"Precit time: {predict_time:.4f} seconds")
    print(f"evaluating: {pipeline['classifier'].__class__.__name__}")
    # Calculate and print the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Model: {pipeline['classifier'].__class__.__name__}\nAccuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))
    print("")
```

```python
def preprocess_text(text):
    # Cleaning: remove special characters and extra spaces
    # text = re.sub(r'[\t\n\r]+', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s.,;\'\":-]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = nltk.word_tokenize(text)

    # Stop Words Removal
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token.lower() not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens]

    # Rejoin tokens into a single string
    return ' '.join(tokens)
```
