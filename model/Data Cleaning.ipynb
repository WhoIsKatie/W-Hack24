```python

import pandas as pd
import numpy
import pickle
infile = open('combined_cleaned.pkl','rb')
df = pickle.load(infile)
df = df.rename(columns={'Email Text': 'input', 'Email Type': 'output'})
df
```

```python
trimmed_df = df.sample(n=100, random_state=37)
```

```python
class_counts = trimmed_df['output'].value_counts()
class_counts
```

```python
trimmed_df
```

```python
trimmed_df = trimmed_df.rename(columns={'input': 'prompt', 'output': 'completion'})
```

```python
trimmed_df
```

```python
prepend_str = """
Given the input email and the reponse determine if the email provided below is safe or potentially malicious based on its
content. If the email seems legitimate and does not contain any
harmful intent, threats, or suspicious requests, classify it as
"safe". If the email contains elements typical of phishing attempts,
such as requests for sensitive information, suspicious links, or unusual
sender addresses, classify it as "malicious".
"""

# Modifying the first column
trimmed_df['prompt'] = trimmed_df['prompt'].apply(lambda x: prepend_str + x)
```

```python
out = trimmed_df.to_json(orient='records')[1:-1].replace('},{', '} {')
with open('finetunedata.json', 'w') as f:
    f.write(out)
```

```python
import json

def reformat_jsonl(input_file_path, output_file_path):
    # Open the original file
    with open(input_file_path, 'r') as infile:
        # Open the output file
        with open(output_file_path, 'w') as outfile:
            # Iterate through each line in the original file
            for line in infile:
                try:
                    # Attempt to load the line as a JSON object
                    data = json.loads(line)
                    # Write the JSON object back to a string ensuring it's in JSONL format
                    jsonl_str = json.dumps(data)
                    outfile.write(jsonl_str + '\n')
                except json.JSONDecodeError as e:
                    print(f"Error processing line: {e}")
                    # Handle lines that can't be decoded as JSON,
                    # e.g., log the error, skip the line, or attempt to fix it.
                    # This example simply skips the line.

# Specify the path to your original file and the path where you want to save the reformatted file
input_file_path = 'finetunedata.json'
output_file_path = 'finetunedata_corrected.jsonl'

# Call the function to reformat the file
reformat_jsonl(input_file_path, output_file_path)
```
